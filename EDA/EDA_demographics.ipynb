## Goal

The goal of this analysis is to figure out the important features that determine the case counts in a given district


## Demographics Data

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
demo_data = pd.read_csv("../Datasets/demographics/Demographics_of_India.csv")
demo_data.shape
```

```python
demo_data['distname']
```

## Case Count dataset

```python
case_data = pd.read_csv("../Datasets/demographics/district_wise.csv")
case_data.shape
```

```python
case_data.head(10)
```

```python
needed_dist = []
for dist in case_data['District']:
#     print(dist)
    if dist in list(demo_data['distname']):
        needed_dist.append(dist)
print("Common districts in both the datasets:", len(needed_dist))
```

Drop the districts in the demographics dataset that are not needed:

```python
demo_data = demo_data.loc[demo_data['distname'].isin(needed_dist)].reset_index(drop=True)
demo_data.shape
```

```python
case_data = case_data.loc[case_data['District'].isin(needed_dist)].reset_index(drop=True)
case_data.shape
```

```python
case_data = case_data.rename(columns={'District':'distname'})
case_data.head()
```

```python
# Merge demographics data and active cases data
district_data = pd.merge(demo_data, case_data, on='distname')
district_data.head()
```

### Make a State wise dataset from district wise dataset

```python
states = set(district_data['statename'])
states = list(states)
states.sort()
```

```python
states_data = district_data.groupby(by='statename').sum()
states_data.head()
```

### Save Datasets

```python
states_data.to_csv("../final_datasets/states_data.csv")
district_data.to_csv("../final_datasets/district_data.csv")
```
